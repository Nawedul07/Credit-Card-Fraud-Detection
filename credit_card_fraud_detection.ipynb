```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

```

```python
# loading the dataset to pandas dataframe
data  = pd.read_csv("/content/creditcard.csv")
data.head(10)
```

```python
data

```

```python
# dataset information
data.info()
```

```python
# checking the number of missing values in each column
data.isnull().sum()
```

```python
# check the distribution of legit and fraudulant transaction
data['Class'].value_counts()
```

label 0 -> legit transactions
label 1 -> fraud transaction

there are only  492 transactions that are fraud, which will lead to underfitting of the model if we fed this data  to the model, so this is a highly unbalanceed dataset

we will have to handle the unblanced data set



```python
# separating the data for analysis
legit_trans = data[data.Class == 0]
fraud_trans = data[data.Class ==1]
print(legit_trans.head())
print(fraud_trans.head())

print(legit_trans.shape)
print(fraud_trans.shape)
```

```python
# statistical measures of the data
legit_trans.Amount.describe()
```

```python
fraud_trans.Amount.describe()
```

```python
#  compare the values for both transaction
data.groupby('Class').mean()
```

Under sampling

build a sample dataset containing similar distribution of normal transaction and the fraudulent transaction

number of fraud transactions is 492


```python
legit_sample = legit_trans.sample(n =492)
legit_sample
```

```python
# concatinating two dataframes
# legit_sample dataset and the fraudulent dataset

```

```python
new_dataset = pd.concat([legit_sample, fraud_trans],axis = 0)
new_dataset.head()
```

```python
new_dataset.shape
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
#  splitting the data into features and targets

X = new_dataset.drop(columns = "Class", axis = 1)
Y = new_dataset["Class"]

print(X)
print(Y)
```

```python
# split the data into training data and test data using train_test_split
# stratify = Y means even split of the fraud and legit transaction in both train and test data

X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.2, stratify = Y, random_state = 2)

print(X.shape)
print(X_train.shape)
print(X_test.shape)
```

```python
# model training
```

```python
# logistic regression for binary classification

model = LogisticRegression()
```

```python
#  traingn logistic regression with training data
model.fit(X_train, Y_train)
```

```python
# model evaluation
```

```python
# Accuracy score on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
print("accuracy of the model on training data : ",training_data_accuracy)
```

```python
# accuracy score on the test data
```

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
print("accuracy on the test data : ", test_data_accuracy)
```

```python
random_sample = legit_trans.sample(n = 1)
print(random_sample)
sample = random_sample.drop(columns = "Class", axis = 1)
print(sample)
```

```python
prediction = model.predict(sample)

if(prediction[0] == 0):
  print("it is a Legit transaction")
else:
  print("It is a Fraud transaction")
```
